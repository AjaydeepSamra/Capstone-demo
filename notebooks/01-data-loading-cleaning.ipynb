```python
#libararies
import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime
```

```python
#loading data, skipping first row because of header
df = pd.read_csv("../data/Gemini_BTCUSD_1h.csv",skiprows=1)
```

```python
#peaking into data
df.head()
```

```python
#looking at columns
df.info()
```

upon looking at the data for the first 51994 rows the unix timestamps were in milliseconds while after that seconds were used

```python
df_tophalf = df.iloc[:51994,:]
df_bottomhalf = df.iloc[51994:,:]
```

```python
#converting float to 64bit int values
df_tophalf['unix'] = df_tophalf['unix'].astype(np.int64)
df_bottomhalf['unix'] = df_bottomhalf['unix'].astype(np.int64)
```

```python
#converting epoch time to local time
df_tophalf['date'] = pd.to_datetime(df_tophalf['unix'], unit='ms')
df_bottomhalf['date'] = pd.to_datetime(df_bottomhalf['unix'], unit='s')
```

```python
df_tophalf.tail()
```

```python
df_bottomhalf.tail()
```

```python
final_df =pd.concat([df_tophalf, df_bottomhalf], axis=0, ignore_index=True)
```

```python
final_df.head(51997)
```

```python
#dropping epoch time because it has been converted, also dropping symbol column because we are focusing on btc first
final_df = final_df.drop(columns=['unix','symbol']).copy()
```

```python
#dropping the last row because it had 0 values
final_df = final_df.drop(index=75823).copy()
```
